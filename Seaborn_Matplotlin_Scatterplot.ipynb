### JointPlot


### Depict Math Score vs Reading Score on a jointplot

```python
sns.jointplot(x=df['Math_Score'],y=df['Reading_Score'],color='blue',kind='scatter')
plt.show()
#  default kind='scatter'
```

### Depict Writing Score vs Reading Score on a jointplot

```python
sns.jointplot(x=df['Writing_Score'],y=df['Reading_Score'],kind='hex',color='red')
plt.show()
```

```python
sns.jointplot(x=df['Writing_Score'],y=df['Reading_Score'],kind='kde',color='green')
plt.show()
# kde = Kernel Density Estimate
```

```python
sns.jointplot(x=df['Writing_Score'],y=df['Reading_Score'],kind='reg',color='orange')
plt.show()
# reg = regression
```
